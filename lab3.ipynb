```
#Bootstarpping
poparray=np.array([1,2,3])
```

```
SampleMean=[]
for _ in range(10000):
  sample=np.random.choice(poparray,2,replace=True)
  SampleMean.append(sample.mean())
SampleMean
```

```
unique,frequency=np.unique(SampleMean,return_counts=True)
unique,frequency
```

```
plt.bar(UniSampleMeanVal,Samplemeanfreq, align='center',width=0.1)
plt.xlabel('Possible sample mean')
plt.ylabel('Frequency')
for i in range(len(Samplemeanfreq)):
  plt.hlines(Samplemeanfreq[i],1,UniSampleMeanVal[i])
plt.show()
```
